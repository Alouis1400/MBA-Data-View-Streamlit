#Storytelling do projeto

Meu chefe que √© um CSO, me pediu para an√°lisar estes dados e trazer valor. As an√°lises feitas foram as seguintes:
* Produtos destaque: top 10 produtos em receita e volume.
* An√°lise geral: vendas totais, ticket m√©dio, quantidade vendida total.
* An√°lise por categoria: quais categorias geram mais receita e volume.
* Sazonalidade: entender como as vendas se comportam ao longo dos meses.


Escolhi essas an√°lises porque d√£o ao CSO uma vis√£o r√°pida e estrat√©gica das vendas, categorias, sazonalidade e produtos-chave. Elas apoiam decis√µes de investimento, estoque e marketing de forma pr√°tica e valiosa.


```python
# An√°lise de Vendas - Loja de Departamento
# 1. Importa√ß√£o das Bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Configurar estilo dos gr√°ficos
sns.set()
```

```python
# 2. Carregamento dos Dados (arquivo j√° no Colab)
df = pd.read_csv('/content/sales_data.csv')
```

```python
# Exibir as primeiras linhas
df.head()
```

```python
# 3. An√°lise Geral
# Vendas Totais
vendas_totais = df['Total_Sales'].sum()
```

```python
# Quantidade Total Vendida
quantidade_total = df['Quantity_Sold'].sum()
```

```python
# Ticket M√©dio
ticket_medio = vendas_totais / quantidade_total
print(f"üîπ Vendas Totais: R$ {vendas_totais:,.2f}")
print(f"üîπ Quantidade Total Vendida: {quantidade_total}")
print(f"üîπ Ticket M√©dio: R$ {ticket_medio:,.2f}")
```

```python
# 4. An√°lise por Categoria
# Receita por Categoria
receita_categoria = df.groupby('Category')['Total_Sales'].sum().sort_values(ascending=False)
```

```python
# Quantidade por Categoria
quantidade_categoria = df.groupby('Category')['Quantity_Sold'].sum().sort_values(ascending=False)
```

```python
# Plot Receita por Categoria (barra horizontal)
plt.figure(figsize=(8,5))
receita_categoria.plot(kind='barh', title='Receita Total por Categoria')
plt.xlabel('Receita (R$)')
plt.tight_layout()
plt.show()
```

```python
# Plot Quantidade por Categoria (barra vertical)
plt.figure(figsize=(8,5))
quantidade_categoria.plot(kind='bar', title='Quantidade Vendida por Categoria')
plt.ylabel('Unidades Vendidas')
plt.xlabel('Categoria')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 5. An√°lise de Sazonalidade (Mensal)
# Ajustar tipo da data
df['Date_Sold'] = pd.to_datetime(df['Date_Sold'])
```

```python
# Agrupar por m√™s usando 'ME'
vendas_mensais = df.resample('ME', on='Date_Sold')['Total_Sales'].sum()
```

```python
# Plot vendas mensais
plt.figure(figsize=(14,7))
ax = vendas_mensais.plot(marker='o', title='Vendas Totais por M√™s')
ax.set_ylabel('Receita (R$)')
ax.set_xlabel('M√™s')
ax.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 6. Top 10 Produtos em Receita e Quantidade
# Top 10 por Receita
top_receita = df.groupby('Product_Name')['Total_Sales'].sum().sort_values(ascending=False).head(10)
```

```python
# Renomear Top Receita
novos_nomes_receita = {
    'Product_90': 'Smartphone X',
    'Product_89': 'Smart TV 55"',
    'Product_93': 'Notebook Pro',
    'Product_80': 'Tablet Ultra',
    'Product_94': 'Fone Bluetooth',
    'Product_78': 'Geladeira 2 Portas',
    'Product_87': 'M√°quina de Lavar',
    'Product_77': 'Micro-ondas Inox',
    'Product_72': 'Console Gamer',
    'Product_71': 'C√¢mera Fotogr√°fica'
}
top_receita = top_receita.rename(index=novos_nomes_receita)
```

```python
# Plot Top Receita com valores
plt.figure(figsize=(8,6))
ax = top_receita.plot(kind='barh', title='Top 10 Produtos por Receita')
for i in ax.patches:
    plt.text(i.get_width() + 50, i.get_y() + 0.4, f'R$ {i.get_width():,.2f}', fontsize=9)
plt.xlabel('Receita (R$)')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
# Top 10 por Quantidade
top_quantidade = df.groupby('Product_Name')['Quantity_Sold'].sum().sort_values(ascending=False).head(10)
```

```python
# Renomear Top Quantidade
novos_nomes_quantidade = {
    'Product_95': 'Headphone Gamer',
    'Product_88': 'Smartwatch Fit',
    'Product_79': 'Cafeteira El√©trica',
    'Product_81': 'Ventilador Turbo',
    'Product_74': 'Batedeira Planet√°ria',
    'Product_96': 'Monitor LED 27"',
    'Product_73': 'Teclado Mec√¢nico',
    'Product_76': 'Aspirador Port√°til',
    'Product_75': 'Liquidificador 900W',
    'Product_91': 'Caixa de Som Bluetooth'
}
top_quantidade = top_quantidade.rename(index=novos_nomes_quantidade)
```

```python
# Plot Top Quantidade com valores
plt.figure(figsize=(8,6))
ax = top_quantidade.plot(kind='barh', title='Top 10 Produtos por Quantidade Vendida')
for i in ax.patches:
    plt.text(i.get_width() + 2, i.get_y() + 0.4, f'{int(i.get_width())}', fontsize=9)
plt.xlabel('Quantidade Vendida')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```
